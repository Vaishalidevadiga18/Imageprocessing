```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
img=cv2.imread(r'C:\Users\VAISHALI\Downloads\CV\download (1).jpg')
if img is None:
    raise FileNotFoundError("the path not found")
gaussian=cv2.GaussianBlur(img,(5,5),0)
median=cv2.medianBlur(img,5)
laplacian=cv2.Laplacian(img,cv2.CV_64F)
laplacian=cv2.convertScaleAbs(laplacian)

titles=['Original Binary','gaussian','median','laplacian']
images=[img,gaussian,median,laplacian]
plt.figure(figsize=(10,10))
for i in range(4):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
