```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(r'C:\Users\VAISHALI\Downloads\CV\sahyadri.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
cv2.waitKey(0)
X, Y, D = img.shape
div1 = X // 2
div2 = Y // 2
top_left = img[:div1, :div2]
top_right = img[:div1, div2:]
bottom_left = img[div1:, :div2]
bottom_right = img[div1:, div2:]
div_img = [top_left, top_right, bottom_left, bottom_right]
titles = ['Top Left', 'Top Right', 'Bottom Left', 'Bottom Right']
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
for i, ax in enumerate(axs.flat):
    ax.imshow(div_img[i])
    ax.set_title(f'{titles[i]}: {div_img[i].shape}')
    ax.axis('on')
plt.tight_layout()
plt.show()
```
