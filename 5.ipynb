```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# Load the image in grayscale mode
image = cv2.imread(r'C:\Users\VAISHALI\Downloads\CV\sahyadri.jpg', cv2.IMREAD_GRAYSCALE)

# Check if image is loaded correctly
if image is None:
    raise FileNotFoundError("The image 'sahyadri.jpg' was not found.")

# Convert the grayscale image to binary using thresholding
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

# Create a 3x3 structuring element for morphological operations
struct_elem = np.ones((3,3), np.uint8)

# Apply erosion - shrinks white regions (foreground)
eroded_image = cv2.erode(binary_image, struct_elem)

# Apply dilation - expands white regions (foreground)
dilated_image = cv2.dilate(binary_image, struct_elem)

# Calculate absolute difference between original and eroded image
# Highlights the pixels lost due to erosion
eroded_diff = cv2.absdiff(eroded_image, binary_image)

# Calculate absolute difference between original and dilated image
# Highlights the pixels gained due to dilation
dilated_diff = cv2.absdiff(dilated_image, binary_image)

# Titles for the plots
titles = ['Original Binary', 'Eroded', 'Dilated', 'Erosion Diff', 'Dilation Diff']
# Images to display
images = [binary_image, eroded_image, dilated_image, eroded_diff, dilated_diff]

# Display the images in a grid using matplotlib
plt.figure(figsize=(16, 6))
for i, img in enumerate(images):
    plt.subplot(2, 3, i+1)
    plt.imshow(img, cmap='gray')  # Display image in grayscale
    plt.title(titles[i])          # Set title
    plt.axis('off')               # Hide axes
plt.tight_layout()
plt.show()
```
